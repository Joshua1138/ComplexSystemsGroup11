```r
library(dplyr)
library(nonlinearTseries)
library(plot3D)

# STEP 1: Compute rate of change
plot_data <- summary_df %>%
  arrange(times_minutes) %>%
  mutate(
    win_gold_change = (lead(mean_gold_w) - lag(mean_gold_w)) / 2,
    lose_gold_change = (lead(mean_gold_l) - lag(mean_gold_l)) / 2,

    win_xp_change = (lead(mean_xp_w) - lag(mean_xp_w)) / 2,
    lose_xp_change = (lead(mean_xp_l) - lag(mean_xp_l)) / 2,

    win_lh_change = (lead(mean_lh_w) - lag(mean_lh_w)) / 2,
    lose_lh_change = (lead(mean_lh_l) - lag(mean_lh_l)) / 2
  ) %>%
  filter(!is.na(win_gold_change), !is.na(lose_gold_change))

# STEP 2: Variables for phase space
win_var <- na.omit(plot_data$win_gold_change)
lose_var <- na.omit(plot_data$lose_gold_change)

# STEP 3: Time lag (tau) estimation
tau.acf_win <- timeLag(win_var, technique = "acf", lag.max = 40, do.plot = TRUE)
tau.ami_win <- timeLag(win_var, technique = "ami", lag.max = 40, do.plot = TRUE)

tau.acf_lose <- timeLag(lose_var, technique = "acf", lag.max = 40, do.plot = TRUE)
tau.ami_lose <- timeLag(lose_var, technique = "ami", lag.max = 40, do.plot = TRUE)

# STEP 4: Embedding dimension estimation
emb_win <- estimateEmbeddingDim(win_var, time.lag = tau.acf_win, max.embedding.dim = 10)
emb_lose <- estimateEmbeddingDim(lose_var, time.lag = tau.acf_lose, max.embedding.dim = 10)

# STEP 5: Phase space reconstruction
ps_win <- buildTakens(win_var, emb_win, tau.acf_win)
ps_lose <- buildTakens(lose_var, emb_lose, tau.acf_lose)

# STEP 6: Plot 3D phase space
lines3D(ps_win[,1], ps_win[,2], ps_win[,3], t = "l", col = "blue", asp = 1)
lines3D(ps_lose[,1], ps_lose[,2], ps_lose[,3], t = "l", col = "red", asp = 1)

# Optional: add legend or title
title("3D Phase Space of Gold Change — Win (Blue) vs Lose (Red)")
```
